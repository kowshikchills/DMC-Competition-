```python
import numpy as np
URL = 'orders_train.txt'
data = open(URL,'r')
lines = data.readlines()
```

```python
#PAYMENT ANALYSIS. 
payment =[]
for i in range(len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    payment.append(T1[-2]) 
payment_array  = np.array(payment[1:])         #payment_diff = number of differnt payment methods 
payment_diff = np.unique(payment_array) 
payment_counts = {}
for i in payment_diff:
    payment_counts[i] = payment.count(i)       #payment_counts = number of people used pirticular payment methods
```

```python
#COLOR ANALYSIS. 
color =[]
for i in range(len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    color.append(T1[3]) 
color_array  = np.array(color[1:])         #color diff = number of differnt colors 
color_diff = np.unique(color_array) 
color_counts = {}
for i in color_diff:
    color_counts[i] = color.count(i)       #color_counts = number of people used pirticular color.counts
```

```python
# quantity ANALYSIS. 
quantity  =[]
for i in range(len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    quantity.append(T1[6])
quantity = [int(i) for i in quantity[1:]]
quantity_array  = np.array(quantity)         #quantity diff = number of differnt quantity 
quantity_diff = np.unique(quantity_array)
quantity_counts = {}
for i in quantity_diff:
    quantity_counts[i] = quantity.count(i) 
```

```python
# rquantity ANALYSIS. 
rquantity  =[]
for i in range(len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    rquantity.append(T1[-1])
rquantity = [int(i) for i in rquantity[1:]]
rquantity_array  = np.array(rquantity)         #quantity diff = number of differnt quantity 
rquantity_diff = np.unique(rquantity_array)
rquantity_counts = {}
for i in rquantity_diff:
    rquantity_counts[i] = rquantity.count(i) 
```

```python
# date ANALYSIS. 
date  =[]
for i in range(len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    date.append(T1[1])
date_array  = np.array(date)         #quantity diff = number of differnt quantity 
date_diff = np.unique(date_array[1:])
date_counts = {}
for i in date_diff:
    date_counts[i] = date.count(i)
```

```python
# color quantity ANALYSIS. 

C_quantity  ={}
for j in color_diff:
    C_quantity[j] = 0  
for i in range(1,len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    C_quantity[T1[3]] =  int(T1[-1]) + C_quantity[T1[3]] 
```

```python
# color order ANALYSIS. 
o_quantity  ={}
for j in color_diff:
    o_quantity[j] = 0  
for i in range(1,len(lines)):
    edata = lines[i]
    T1 = tuple(filter(None, edata.split(';')))
    o_quantity[T1[3]]  =  int(T1[6]) + o_quantity[T1[3]]
```

```python
#proba color reject
prob_color_reject = {}
for i in color_diff:
    prob_color_reject[i] = float(C_quantity[i])/o_quantity[i]
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer
import pickle
```

```python
df_train = pd.read_csv('orders_train.txt', sep=';')

df_train.productGroup = df_train.productGroup.astype(str)
df_train.colorCode = df_train.colorCode.astype(str)
df_train.deviceID = df_train.deviceID.astype(str)
df_test = pd.read_csv('orders_class.txt', sep=';')
df_test.productGroup = df_test.productGroup.astype(str)
df_test.colorCode = df_test.colorCode.astype(str)
df_test.deviceID = df_test.deviceID.astype(str)


y = np.array(df_train.returnQuantity) 

df_train = df_train.loc[:,'articleID':'returnQuantity']
df_test.loc[:,'articleID':]
df_train = df_train.fillna('51')
df_test = df_test.fillna('51')
df_train.customerID = 0
df_test.customerID = 0

X_dict = df_train.to_dict('records')
X_test_dict = df_test.to_dict('records')

X_en = []
X_en.extend(X_dict)
X_en.extend(X_test_dict)
# Encoded SET, Encoding both TESTING AND TRAINING SET
#filling NAN values as '51'
```

```python
df_train
```

```python
#One Hot Encoding 
enc = DictVectorizer()
X_encoded = enc.fit_transform(X_en)
```

```python
X_encoded
```

```python
#Validating  SET
train_split = int(0.25*df_train.shape[0])
X_vald =  X_encoded[0:train_split]
Y_vald = y[0:train_split]
#Training SET
test_split = len(X_dict)
X_train = X_encoded[train_split:test_split]
Y_train = y[train_split:test_split]
#Testing SET
X_test =  X_encoded[test_split:]
#Orginal Model SET, Complete data for Training 
X = X_encoded[:train_split]
Y = y[:train_split]
```

```python
pickle.dump([X_vald,Y_vald, X_train,Y_train, X_test,X,Y], open('megadata2.pkl','wb'))
```
